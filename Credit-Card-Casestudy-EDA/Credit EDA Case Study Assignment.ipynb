<font color='blue'>**Estimating Bins**</font>

````python
# Estimating Bins  
print(final_df['DAYS_EMPLOYED'].max()/365 ,final_df['DAYS_BIRTH'].max()/365 ,final_df['DAYS_LAST_PHONE_CHANGE'].max()/365 ,final_df['DAYS_REGISTRATION'].max()/365 )
````

````python
def binning(n,r):
 bins=[]
 labels=[]
 for i in range(0,r):
    bins.append(365*i*n)
    s=str(i*n)+'-'+str((i+1)*n)
    labels.append(s)
 del labels[-1]   # deleting last label as  Bin labels must be one fewer than the number of bin edges/''
 return bins,labels


# for DAYS_REGISTRATION & DAYS_BIRTH columns Creating Bins of 10 years upto 80  
bins,labels=binning(10,8)
final_df['DAYS_BIRTH_Bins']=pd.cut(final_df['DAYS_BIRTH'], bins=bins, labels=labels)
final_df['DAYS_REGISTRATION_Bins']=pd.cut(final_df['DAYS_REGISTRATION'], bins=bins, labels=labels)

# for DAYS_LAST_PHONE_CHANGE column Creating Bins of 1 years upto 12years
bins,labels=binning(1,12)
final_df['DAYS_LAST_PHONE_CHANGE_Bins']=pd.cut(final_df['DAYS_LAST_PHONE_CHANGE'], bins=bins, labels=labels)
# for DAYS_EMPLOYED column creating bins of 5 years upto 50 years 
bins,labels=binning(5,10)
bins.insert(0,-1) # added a extra bin to indentify umemployed
bins[1]=1
labels.insert(0,'0')
labels[1]='1-5'
final_df['DAYS_EMPLOYED_Bins']=pd.cut(final_df['DAYS_EMPLOYED'], bins=bins, labels=labels)
final_df['DAYS_LAST_PHONE_CHANGE_Bins']=final_df['DAYS_LAST_PHONE_CHANGE_Bins'].replace(np.nan, 0)
````
